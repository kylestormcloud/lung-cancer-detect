```python
import SimpleITK as sitk
import matplotlib.pylab as plt
f = "D:\datasets\LUNA"
os.chdir(f)
one_mhd = "1.3.6.1.4.1.14519.5.2.1.6279.6001.105756658031515062000744821260.mhd"
ct_scans = sitk.GetArrayFromImage(sitk.ReadImage(one_mhd, sitk.sitkFloat32))
print("Total images in series =  {}  for file: {}".format(ct_scans.shape[0],one_mhd))
```

```python
plt.figure(figsize=(20,16))
plt.gray()
plt.subplots_adjust(left = 0, right = 1, bottom = 0, top = 1, wspace = .1, hspace = .1)
for i in range(ct_scans.shape[0]):
    plt.subplot(12,11,i+1), plt.imshow(ct_scans[i]), plt.axis('off')
plt.show()
```
